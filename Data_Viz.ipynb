## Level Up: Data Analysis Example & Using Seaborn

```python
#loads tips dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# visualize the relationship between bill and tip

# your code here
bill_tip = tips[['total_bill', 'tip']]

bill = tips.total_bill
tip = tips.tip

fig, ax = plt.subplots()

ax.scatter(bill, tip)
ax.set_title("Bill & Tip");
```

```python
# visualize the average tips men vs women pay - what's the appropriate plot?

sex_tip = tips[['sex', 'tip']]

male = sex_tip[sex_tip['sex']=='Male']
female = sex_tip[sex_tip['sex']=='Female']

avg_male = sum(male['tip']) / len(male['tip'])

avg_female = sum(female['tip']) / len(female['tip'])

label = ['Male', 'Female']

avg_tip = [avg_male, avg_female]

fig, ax = plt.subplots()

ax.bar(label, avg_tip);
```

```python
# your turn - visualize the average amount of total bill for time (lunch or dinner)

time_bill = tips[['time', 'total_bill']]

lunch = time_bill[time_bill['time'] == 'Lunch']
dinner = time_bill[time_bill['time'] == 'Dinner']

avg_lunch = sum(lunch['total_bill']) / len(lunch['total_bill'])
avg_dinner = sum(dinner['total_bill']) / len(dinner['total_bill'])

label = ['Lunch', 'Dinner']
avg_time = [avg_lunch, avg_dinner]

fig, ax = plt.subplots()
ax.bar(label, avg_time);
```
